### The following is a fix for legacy taxonomy codes that are still in the data:  The master list they provided does not have some old taxonomy codes.  The code would break if we have labels that are not in the dictionary.

### So with the conversion chart that business provided we go through and convert any old codes in our data to be trained/tested.

```python
tcc_2018 = pd.read_csv('tcc_2018.csv')
tcc = tcc_2018.loc[:41,:]
#sas_16_1_test[sas_16_1_test['taxonomy_id'].str.contains('61P12401')] = '61P12105'

import pickle
pkl_file = open('master_dict_20pct_rs1', 'rb')
master_dict = pickle.load(pkl_file)

#load data file
import pickle
pkl_file = open('ss_20pct_18_1_str_ID.pkl', 'rb')
sas_sample = pickle.load(pkl_file)

all_tax_ids = pd.read_csv('all_tax_ids.csv', names=['VERIZON CODE'])
#First encode the response variable taxonomy_id as Y
from sklearn.preprocessing import LabelEncoder
encoder = LabelEncoder()
Y = all_tax_ids['VERIZON CODE']
encoder.fit(Y)

#Apply the conversion
tc_dict = dict(zip(tcc['TaxonomyID_Old'], tcc['New']))
for x in tc_dict:
    sas_sample['taxonomy_id'].replace(x, tc_dict[x],inplace=True)

#extract features and target Y
sas_features = sas_sample[list(master_dict.keys())]

#first transform Y: a couple of pesky codes that remains?
#sas_sample[sas_sample['taxonomy_id'].str.contains('61P12401')] = '61P12105'
#sas_sample[sas_sample['taxonomy_id'].str.contains('61P12402')] = '61P12105'

Y = sas_sample['taxonomy_id']
encoded_Y = encoder.transform(Y)
```

## The following code makes sure that any new labels in the new data are incorporated into the master dictionary to ensure that our application is robust against new entries

```python
t1 = time.time()
#now process X: deal with new labels, encode, normalize
X = np.zeros(sas_features.shape)

#Encode with our master dictionary
i = 0
comb_codes = dict()

for col in master_dict.keys():
    #These lines below allows incorporating newer lables into master dictionary
    comb_col = [master_dict[col].classes_, sas_features[col]]
    merged = list(set().union(*comb_col))
    comb_codes[col] = merged
    master_dict[col] = LabelEncoder().fit(comb_codes[col])
    #Now label encode the data set
    X[:,i] = master_dict[col].transform(sas_features[col])
    i += 1
    #print(col, 'is now processed')

t2 = time.time()
print('This part of proprocessing took', t2-t1, 'secs of time')
```

## Now define the model

```python
# define model - adjut dimensions of the dataset depending which one is used
def basic_model():
    optimizer = 'rmsprop'
    input_size = X_train.shape[0]
    input_dim = X_train.shape[1]
    output_dim = Y_train.shape[1]
    activation = 'relu'
    # create model
    model = Sequential()
    #This is a hidden layer with input_size many activations?
    #model.add(Dense(input_size, input_dim=input_dim, activation=activation))
    #This is a hidden layer with input_size many activations?
    model.add(Dense(500, input_dim=input_dim, activation=activation))
    #model.add(Dropout(0.3))
    model.add(Dense(output_dim, activation='softmax'))
    # Compile model
    #sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
    model.compile(loss='categorical_crossentropy', optimizer=optimizer, metrics=['categorical_accuracy'])
    
    return model

###https://github.com/keunwoochoi/keras_callbacks_example/blob/master/my_callbacks.py
```

```python
#Now the model
model = KerasClassifier(build_fn=basic_model, epochs=10, batch_size=128,  validation_split=0.1, verbose=1)
```
